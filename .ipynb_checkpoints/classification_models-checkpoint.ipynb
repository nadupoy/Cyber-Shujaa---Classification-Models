### 3. Prepare the data:

- Split the dataset into training and test sets.

```python
# Define Features (X): Drop the 'target', column to get the input data:
X = wine_dataset_df.drop('target', axis=1)

# Define Target (y): Select only the 'target' column:
y = wine_dataset_df['target']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.4,    # 40% of data for testing
    random_state=42,  # Ensures reproducibility
    stratify=y        # Recommended for classification to keep class balance
)

# Scale and transform the training data using the StandardScaler from the preprocessing module:
scaler_train = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = scaler_train.transform(X_train)

# Scale and transform the test data using the StandardScaler from the preprocessing module:
scaler_test = preprocessing.StandardScaler().fit(X_test)
X_test_scaled = scaler_test.transform(X_test)
```

```python
# Scaled data has zero mean and unit variance:
X_train_scaled.mean(axis=0)
```

```python
X_train_scaled.std(axis=0)
```

### 4. Build the models:

- Logistic Regression:

```python
# Apply the Logistic Regression model:
# Instantiation:
logistic_regression = linear_model.LogisticRegression(
    solver='lbfgs',       # Default solver, good for small datasets
    max_iter=1000
)

# Training/Fitting:
logistic_regression = logistic_regression.fit(X_train_scaled, y_train)
logistic_regression
```

- Decision Tree:

```python
# Apply the Decision Tree model (all data has to be numerical):
decision_tree = tree.DecisionTreeClassifier()
decision_tree = decision_tree.fit(X_train_scaled, y_train)

tree.plot_tree(
    decision_tree, 
    feature_names=X.columns.tolist(),
    filled=True,                         # Optional: adds colour for clarity
    rounded=True                         # Optional: rounds box corners
)
plt.show()
```

- Random Forest:

```python
# Apply the Random Forest model:
random_forest = ensemble.RandomForestClassifier(
    max_depth=5,
    n_estimators=10,
    max_features=1,
    random_state=42
)
random_forest = random_forest.fit(X_train_scaled, y_train)
```

- K-Nearest Neighbours (KNN):

```python
# Apply the K-Nearest Neighbours (KNN) model:
k_nearest_neighbors = neighbors.KNeighborsClassifier(3)
k_nearest_neighbors = k_nearest_neighbors.fit(X_train_scaled, y_train)
```

- Naive Bayes:

```python
naive_bayes = GaussianNB()
naive_bayes = naive_bayes.fit(X_train_scaled, y_train)
```

- Support Vector Machine (SVM):

```python
support_vector_machine = svm.SVC(
    kernel="linear",
    C=0.025,
    random_state=42
)
support_vector_machine = support_vector_machine.fit(X_train_scaled, y_train)
```

### 5. Evaluate the models:

- Accuracy score:

```python
# 1. Make predictions on the test set features:
y_pred_logistic_regression = logistic_regression.predict(X_test_scaled)

# 2. Calculate the accuracy by comparing predictions to true labels:
accuracy_logistic_regression = accuracy_score(y_test, y_pred_logistic_regression)

print(f"Logistic Regression Model Accuracy: {accuracy_logistic_regression:.4f}")
```

```python
y_pred_decision_tree = decision_tree.predict(X_test_scaled)
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)

print(f"Decision Tree Model Accuracy: {accuracy_decision_tree:.4f}")
```

```python
y_pred_random_forest = random_forest.predict(X_test_scaled)
accuracy_random_forest = accuracy_score(y_test, y_pred_random_forest)

print(f"Random Forest Model Accuracy: {accuracy_random_forest:.4f}")
```

```python
y_pred_k_nearest_neighbors = k_nearest_neighbors.predict(X_test_scaled)
accuracy_k_nearest_neighbors = accuracy_score(y_test, y_pred_k_nearest_neighbors)

print(f"K-Nearest Neighbors Model Accuracy: {accuracy_k_nearest_neighbors:.4f}")
```

```python
y_pred_naive_bayes = naive_bayes.predict(X_test_scaled)
accuracy_naive_bayes = accuracy_score(y_test, y_pred_naive_bayes)

print(f"Naive Bayes Model Accuracy: {accuracy_naive_bayes:.4f}")
```

```python
y_pred_support_vector_machine = support_vector_machine.predict(X_test_scaled)
accuracy_support_vector_machine = accuracy_score(y_test, y_pred_support_vector_machine)

print(f"Support Vector Machine Model Accuracy: {accuracy_support_vector_machine:.4f}")
```

- Classification report:

```python
# Logistic Regression:
print("--- Classification Report: Logistic Regression ---")
print(classification_report(y_test, y_pred_logistic_regression))
```

```python
# Decision Tree:
print("--- Classification Report: Decision Tree ---")
print(classification_report(y_test, y_pred_decision_tree))
```

```python
# Random Forest:
print("--- Classification Report: Random Forest ---")
print(classification_report(y_test, y_pred_random_forest))
```

```python
# K-Nearest Neighbors:
print("--- Classification Report: K-Nearest Neighbors ---")
print(classification_report(y_test, y_pred_k_nearest_neighbors))
```

```python
# Naive Bayes:
print("--- Classification Report: Naive Bayes ---")
print(classification_report(y_test, y_pred_naive_bayes))
```

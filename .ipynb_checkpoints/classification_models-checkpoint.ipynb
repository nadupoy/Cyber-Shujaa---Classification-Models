```python
# Import required libraries:
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_wine

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

### 1. Load the dataset:

- Load and explore the wine dataset from `sklearn.datasets`

```python
# Load the data, returning the entire dataset as a single Bunch object containing a 'frame' attribute (the DataFrame):
wine_dataset_bunch = load_wine(return_X_y=False, as_frame=True)

# Access the DataFrame directly from the 'frame' attribute:
wine_dataset_df = wine_dataset_bunch.frame
wine_dataset_df
```

```python
# Print a concise summary of the DataFrame:
wine_dataset_df.info()
```

```python
# Generate descriptive statistics:
wine_dataset_df.describe().T
```

```python
# Generate the mean of the values in each column:
wine_dataset_df.mean()
```

```python
# Return the sample standard deviation of each column:
wine_dataset_df.std()
```

```python
# Return the median of the values in each column i.e. the 50th percentile:
wine_dataset_df.median()
```

### 2. Explore the dataset:

- Perform *exploratory data analysis* (EDA) to understand feature relationships and class distributions.

Distribution of values in each column:

```python
# Box plot showing distribution of values in each DataFrame column:
fig, ax = plt.subplots()    # a figure with a single axes
ax.set_title("Chemical Composition", pad=20.0)
ax.set_ylabel("Quantity", labelpad=10.0)
ax.set_xlabel("Wine Characteristics", labelpad=10.0)
ax.boxplot(wine_dataset_df)

plt.show()
```
